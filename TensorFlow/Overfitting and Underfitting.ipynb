```python
import tensorflow as tf
from tensorflow import keras
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
%matplotlib inline
```

```python
print(tf.__version__)
```

### IMDB dataset

```python
imdb= keras.datasets.imdb
```

```python
(X_train, y_train), (X_test, y_test)= imdb.load_data(num_words=10000)
```

#### Multi hot encoding

```python
def multi_hot_sequence(sequences, dim):
    result= np.zeros((len(sequences), dim))
    for i, word_index in enumerate(sequences):
        result[i, word_index]= 1.0
    return result
```

```python
X_train= multi_hot_sequence(X_train, dim= 10000)
X_test= multi_hot_sequence(X_test, dim= 10000)
```

```python
plt.plot(X_train[45])
```

### Trying different model size

#### Baseline model

```python
baseline_model= keras.Sequential([
    keras.layers.Dense(16, activation=tf.nn.relu, input_shape= (10000,)),
    keras.layers.Dense(16, activation=tf.nn.relu),
    keras.layers.Dense(1, activation=tf.nn.sigmoid)
])
```

```python
baseline_model.compile(optimizer='adam',
                      loss= 'binary_crossentropy',
                      metrics= ['accuracy', 'binary_crossentropy'])
```

```python
baseline_model.summary()
```

```python
baseline_history= baseline_model.fit(X_train,
                                    y_train,
                                    epochs= 20,
                                    batch_size= 512,
                                    validation_data= (X_test, y_test),
                                    verbose= 2)
```

#### Smaller model

```python
smaller_model= keras.Sequential([
    keras.layers.Dense(4, activation= tf.nn.relu, input_shape= (10000,)),
    keras.layers.Dense(4, activation= tf.nn.relu),
    keras.layers.Dense(1, activation= tf.nn.sigmoid)
])
```

```python
smaller_model.compile(optimizer='adam',
                     loss= 'binary_crossentropy',
                     metrics= ['accuracy', 'binary_crossentropy'])
```

```python
smaller_model.summary()
```

```python
smaller_history= smaller_model.fit(X_train, 
                                  y_train,
                                  validation_data= (X_test, y_test),
                                  epochs= 20,
                                  batch_size= 512,
                                  verbose= 2)
```

#### Bigger model

```python
bigger_model= keras.Sequential([
    keras.layers.Dense(512, activation= tf.nn.relu, input_shape= (10000,)),
    keras.layers.Dense(512, activation= tf.nn.relu),
    keras.layers.Dense(1, activation= tf.nn.sigmoid)
])
```

```python
bigger_model.compile(optimizer='adam',
                     loss= 'binary_crossentropy',
                     metrics= ['accuracy', 'binary_crossentropy'])
```

```python
bigger_model.summary()
```

```python
bigger_history= bigger_model.fit(X_train, 
                                  y_train,
                                  validation_data= (X_test, y_test),
                                  epochs= 20,
                                  batch_size= 512,
                                  verbose= 2)
```

```python
def plot_histories(histories, key= 'binary_crossentropy'):
    plt.figure(figsize=(16,10))
    
    for name, history in histories:
        val= plt.plot(history.epoch, history.history['val_'+key], '--', label= name.title()+" Val")
        plt.plot(history.epoch, history.history[key], color= val[0].get_color(), label= name.title()+" Train")
        
        plt.xlabel("Epochs")
        plt.ylabel(key.replace('_',' ').title())
        plt.legend()
        
        plt.xlim([0, max(history.epoch)])
```

```python
plot_histories([('Smaller', smaller_history), 
               ('Baseline', baseline_history),
               ('Bigger', bigger_history)],
              key= 'binary_crossentropy')
```

### Weight Regularization

```python
l2_model= keras.Sequential([
    keras.layers.Dense(16, kernel_regularizer= keras.regularizers.l2(0.001),
                      activation= tf.nn.relu, input_shape= (10000, )),
    keras.layers.Dense(16, kernel_regularizer= keras.regularizers.l2(0.001),
                      activation= tf.nn.relu),
    keras.layers.Dense(1, activation= tf.nn.sigmoid)
])
```

```python
l2_model.compile(optimizer= 'adam',
                loss= 'binary_crossentropy',
                metrics= ['accuracy', 'binary_crossentropy'])
```

```python
l2_history= l2_model.fit(X_train,
                        y_train,
                        epochs= 20,
                        batch_size= 512,
                        validation_data= (X_test, y_test),
                        verbose= 2)
```

```python
plot_histories([
    ('Baseline', baseline_history),
    ('L2', l2_history)
])
```

### Dropout

```python
dpt_model= keras.Sequential([
    keras.layers.Dense(16, activation= tf.nn.relu, input_shape= (10000,)),
    keras.layers.Dropout(0.5),
    keras.layers.Dense(16, activation= tf.nn.relu),
    keras.layers.Dropout(0.2),
    keras.layers.Dense(1, activation= tf.nn.sigmoid)
])
```

```python
dpt_model.compile(optimizer= 'adam',
                 loss= 'binary_crossentropy',
                 metrics= ['accuracy', 'binary_crossentropy'])
```

```python
dpt_history= dpt_model.fit(X_train, y_train,
                          validation_data= (X_test, y_test),
                          epochs= 20,
                          batch_size= 512,
                          verbose= 2)
```

```python
plot_histories([
    ('Baseline', baseline_history),
    #('L2', l2_history),
    ('Dropout', dpt_history),
])
```

```python
plot_histories([
    ('Baseline', baseline_history),
    ('L2', l2_history),
    ('Dropout', dpt_history),
])
```
